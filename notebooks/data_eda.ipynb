```python
# Create histograms for all 6 columns
import matplotlib.pyplot as plt
df.hist(bins=50, figsize=(20, 15))
plt.show()
# Create scatter plots for price vs each of the other columns
import seaborn as sns
plt.figure(figsize=(20, 15))
sns.scatterplot(x='Area', y='Price', data=df)

# Average price per Bathroom value
avg_price_per_bathroom = df.groupby('Bathrooms')['Price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Bathrooms', y='Price', data=avg_price_per_bathroom)

# Average price per Bedroom value
avg_price_per_bedroom = df.groupby('Bedrooms')['Price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Bedrooms', y='Price', data=avg_price_per_bedroom)

# Average price per Floor value
avg_price_per_floor = df.groupby('Floors')['Price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Floors', y='Price', data=avg_price_per_floor)

# Average price per Built Year
avg_price_per_built_year = df.groupby('YearBuilt')['Price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='YearBuilt', y='Price', data=avg_price_per_built_year)


```

```python
df['bedroom_bathroom_ratio'] = df['Bedrooms'] / df['Bathrooms']
# Create a scatter plot for bedroom_bathroom_ratio vs Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='bedroom_bathroom_ratio', y='Price', data=df)

# Mean price per bedroom_bathroom_ratio
mean_price_per_ratio = df.groupby('bedroom_bathroom_ratio')['Price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='bedroom_bathroom_ratio', y='Price', data=mean_price_per_ratio)

# Mean price by Condition variable
mean_price_per_condition = df.groupby('Condition')['Price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Condition', y='Price', data=mean_price_per_condition)

# Mean price by location
mean_price_per_location = df.groupby('Location')['Price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Location', y='Price', data=mean_price_per_location)

# Mean price by Garage variable
mean_price_per_garage = df.groupby('Garage')['Price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Garage', y='Price', data=mean_price_per_garage)
```
