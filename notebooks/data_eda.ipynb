```python
# Create histograms for all 6 columns
import matplotlib.pyplot as plt
df.hist(bins=50, figsize=(20, 15))
plt.show()
# Create scatter plots for price vs each of the other columns
import seaborn as sns
plt.figure(figsize=(20, 15))
sns.scatterplot(x='Area', y='Price', data=df)

# Average price per Bathroom value
avg_price_per_bathroom = df.groupby('Bathrooms')['Price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Bathrooms', y='Price', data=avg_price_per_bathroom)

# Average price per Bedroom value
avg_price_per_bedroom = df.groupby('Bedrooms')['Price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Bedrooms', y='Price', data=avg_price_per_bedroom)

# Average price per Floor value
avg_price_per_floor = df.groupby('Floors')['Price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Floors', y='Price', data=avg_price_per_floor)

# Average price per Built Year
avg_price_per_built_year = df.groupby('YearBuilt')['Price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='YearBuilt', y='Price', data=avg_price_per_built_year)


```
